```python
def getburstAv(model1,neuronID,threshold):
    #threshold = 0.01
    spikes = model1.params['neurons'][neuronID]['spikeOuts']
    lastSpikeTime = 0.0
    burstOn = False
    lastBurstStart = 0.0
    bursts = []
    diffs = []
    noOfSpikesPB = 0
    for thisSpike in spikes:
        diff = thisSpike-lastSpikeTime
        if diff > threshold:
            burstOnNew = False
        else:
            burstOnNew = True
            noOfSpikesPB +=1
            if burstOn==True:
                diffs.append(diff)
        if burstOnNew==True and burstOn==False:
            lastBurstStart = lastSpikeTime
            noOfSpikesPB = 1
        if burstOnNew==False and burstOn==True:
            bursts.append([lastBurstStart,lastSpikeTime,lastSpikeTime-lastBurstStart,noOfSpikesPB])
        lastSpikeTime = thisSpike
        burstOn = burstOnNew
    burstInfo = []
    x = []
    spk = []
    for burst in bursts:
        x.append(burst[2])
        spk.append(burst[3])
    burstInfo = [np.median(x),np.median(spk),np.mean(diffs),len(bursts),np.mean(x)]
    return burstInfo
# median burst length, median number of spikes, mean difference between spikes and number of bursts
```
